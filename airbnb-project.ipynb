```python
elysee_listings_accommodates = paris_listings.query("neighbourhood=='Elysee'").groupby("accommodates").agg({"price":"mean"}).sort_values("price",ascending = False)
# avg price of per accommodates in neighbourhood "Elysee"  # sampleing 
```

```python
plt.figure(figsize = (12,8))
sns.barplot(data = paris_listings_neighbourhood,
           x = "price",
           y = paris_listings_neighbourhood.index,
           color = "#BDE038")
plt.gca().set_facecolor("#F0F2F0")
plt.gcf().set_facecolor("#F0F2F0")
plt.title("Neighbourhood i Paris(Premium Localities at the Top!)",fontweight='bold',fontsize=12)
plt.xlabel("How Expensive? (Euros)",fontweight='bold',fontsize=12)
plt.ylabel("Prominent places with AirBnBs",fontweight='bold',fontsize=12)
```

```python
plt.figure(figsize= (12,8))
sns.barplot(data = elysee_listings_accommodates,
           x = "price",
           y = elysee_listings_accommodates.index,
           color = "#F28972",
           orient = "h",
           order = elysee_listings_accommodates.index)
plt.gca().set_facecolor("#F0F2F0")
plt.gcf().set_facecolor("#F0F2F0")
plt.title("Average price of AirBnB in Elysee -- vs -- number of people that can stay",fontweight='bold',fontsize=12)
plt.xlabel("Average price (Euros)",fontweight='bold',fontsize=12)
plt.ylabel("Number of people that can stay",fontweight='bold',fontsize=12)
```

```python
plt.figure(figsize = (12,8),facecolor='#f0f0f0')
ax = sns.lineplot(data = paris_listings_over_time["neighbourhood"],color = "green", marker='o',markerfacecolor='green',linewidth=3)
plt.gca().set_facecolor("#F0F2F0")
plt.gcf().set_facecolor("#F0F2F0")
plt.xlabel("Years",fontweight='bold',fontsize=13)
plt.ylabel("Number of AirBnBs and Localities",fontweight='bold',fontsize=13)
plt.title("Popularity of AirBnBs over time",fontweight='bold')
plt.grid(False, color="#545353")
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.lineplot(data = paris_listings_over_time["price"],color = "#075AD9",marker='D',linewidth=3)
plt.xlabel("Years",fontweight='bold',fontsize=12)
plt.ylabel("Average Price",fontweight='bold',fontsize=12)
plt.title("Average Price of AirBnBs over time",fontweight='bold',fontsize=12)
plt.gca().set_facecolor("#F0F2F0")
plt.gcf().set_facecolor("#F0F2F0")
plt.grid(True)
plt.show()
```

<div style="background-color: #FFFFFF; padding: 25px; border-radius: 10px; font-size: 20px; font-weight: 10px; text-align: center;">
  <h1 style="color: #E7455E;"><strong>Finding effects of
      Regulations on AirBnB Business</strong></h1>
</div>

```python
fig , ax = plt.subplots()
ax.plot(paris_listings_over_time.index,paris_listings_over_time["neighbourhood"], label = "New Hosts", c = "green",marker='o',linewidth=2)
for line in ax.lines:
    line.set_markersize(5)
ax.set_ylabel("New Hosts",fontweight='bold',fontsize=11)
ax2 = ax.twinx()
ax2.plot(paris_listings_over_time.index,paris_listings_over_time["price"], label = "Average Price",marker='o',linewidth=2)
for line in ax2.lines:
    line.set_markersize(5) 
    
plt.gca().set_facecolor("#F0F2F0")
plt.gcf().set_facecolor("#F0F2F0")
ax2.set_ylim(0)
ax2.set_ylabel("Average Price",fontweight='bold',fontsize=11)
ax.set_title("Regulations Lead to Fewer New Hosts and Higher Prices",fontweight='bold',fontsize=11)
plt.grid(color="black")
plt.show()
```

<div style="background-color: #FFFFFF; padding: 25px; border-radius: 10px; font-size: 20px; font-weight: 10px; text-align: center;">
  <h1 style="color: #E7455E;"><strong>Final Insights</strong></h1>
</div>

### 1. The Number of AirBnBs kept on increasing since the launch and prices kept on increasing too, due to initial traction and early adopters.
### 2. After the startup is known to everyone and becomes a common utility, AirBnBs start increasing in numbers and prices also kept on decreasing.
### 3. After regulation was announced around 2015 there was under confidence in the business, number of AirBnBs started decreasing and prices               started       increasing.
### 4. Once the regulation is the new normal, during the year 2019 the number of AirBnBs have increased in number and prices kept decreasing due to more   supply of them.

<div style="background-color: #FFFFFF; padding: 25px; border-radius: 10px; font-size: 20px; font-weight: 10px; text-align: center;">
  <h1 style="color: #E7455E;"><strong>Recommendation for AirBnB</strong></h1>
</div>

### 1. Regulations in long term rentals can impact the business adversely, there might be customer and hosts churn due to uncertainty.
### 2. Such regulations might add to AirBnBs losses which might be difficult to recover later.
### 3. If the customer experience is going to get impacted due to this, it would lead to incorrect brand perception.
### 4. It is recommended to watch out for any such regulations at other places and be prepared for it.
### 5. AirBnB can replicate such regulations at other places.
### 6. They can keep strict rules to onboard and release the hosts.
### 7. They can limit the number of AirBnBs in a locality to ensure the public has enough rental options and the government doesn't step in.
